```python
import pandas as pd
from pathlib import Path
from tabulate import tabulate
```

```python
folder_path = "unordered"
```

```python
#Read & Show Info Table 

directory = Path(folder_path)
file_names = [f.name for f in directory.iterdir() if f.is_file()]
df_dict = {'main.csv':None} #至少要有 'main.csv'

for file in file_names:
    if '.csv' in file:
        df = pd.read_csv(f"{folder_path}\{file}")
        df_dict[file] = df

for key in df_dict.keys():
    print( key )

    df = df_dict[key]

    summary = pd.DataFrame({
    "column_name": df.columns,
    "data_type": df.dtypes.values,
    "missing_count": df.isnull().sum().values,
    "missing_percent": (df.isnull().mean() * 100).round(2).values
    })

    print(tabulate(summary, headers="keys", tablefmt="pretty"))
```
